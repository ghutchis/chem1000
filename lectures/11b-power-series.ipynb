### Reminder - Power Series

In Chapter 7, we discussed using a *power series* as a polynomial approximation to a function:

<div class="alert alert-block alert-success">

Many functions can be represented by an infinite series in powers of $x$:

$$
S(x)=\sum_{n=0}^{\infty} a_{n} x^{n}=a_{0}+a_{1} x+a_{2} x^{2}+\ldots
$$
    
</div>

In particular, we saw that it's very easy to take derivatives or integrals of power series because they're just polynomials:

For derivatives:

$$
\begin{aligned}
S(x) &=\left.\frac{d}{d t} f(t)\right|_{t=x}=\left.\frac{d}{d t} \sum_{n=0}^{\infty} a_{n} t^{n}\right|_{t=x}=\sum_{n=0}^{\infty}\left(\left.\frac{d}{d t} a_{n} t^{n}\right|_{t=x}\right) \\
&=\sum_{n=0}^{\infty} a_{n} n x^{n-1}
\end{aligned}
$$

And for integrals

$$
\begin{aligned}
S(x) &=\int_{0}^{x} f(t) d t=\int_{0}^{x}\left(\sum_{n=0}^{\infty} a_{n} t^{n}\right) d t=\sum_{n=0}^{\infty}\left(\int_{0}^{x} a_{n} t^{n} d t\right) \\
&=\sum_{n=0}^{\infty} \frac{1}{n+1} a_{n} x^{n+1}
\end{aligned}
$$

Perhaps we can use power series to help us solve more complicated differential equations...

### First Order Differential Equations with Power Series

Let's start with an easy example:

$$
\frac{d}{d x} y(x)+y(x)=0
$$

**HINT: Solve this first using another approach**

This time, let's use a different *ansatz*

$$
y(x)=\sum_{n=0}^{\infty} a_{n} x^{n}
$$

So we'll need to plug this in and take the derivative:

$$
\frac{d}{d x} y(x)=\frac{d}{d x} \sum_{n=0}^{\infty} a_{n} x^{n}=\sum_{n=0}^{\infty} a_{n}\left(\frac{d}{d x} x^{n}\right)=\sum_{n=0}^{\infty} a_{n} n x^{n-1}
$$

So now we'd have the equation:

$$
\frac{d}{d x} y(x)+y(x)=\sum_{n=0}^{\infty} a_{n} n x^{n-1}+\sum_{n=0}^{\infty} a_{n} x^{n}=0
$$

The big problem is that we have different exponents of $x^n$ and $x^{n-1}$ so we can't immediately combine the two series.

It usually helps to write out the terms of the new series - in this case, after taking the derivative.

$$
    a_0 \times 0 x^{-1} + a_{1}\times 1 x^{0} + a_{2} \times 2 x + a_{3} \times 3 x^{2} \ldots
$$

Clearly the $a_0$ term drops out:

$$
a_{1}\times 1 x^{0} + a_{2} \times 2 x + a_{3} \times 3 x^{2} \ldots
$$

We started with $n=0$ so we want to write this series for the derivative starting at $n=0$ too:

$$
\sum_{n=1}^{\infty} a_{n} n x^{n-1} \rightarrow \sum_{n=0}^{\infty} a_{n+1}(n+1) x^{n}
$$

Our differential equation now looks something like this:

$$
\begin{aligned}
\frac{d}{d x} y(x)+y(x) &=\sum_{n=0}^{\infty} a_{n+1}(n+1) x^{n}+\sum_{n=0}^{\infty} a_{n} x^{n} \\
&=\sum_{n=0}^{\infty}\left[a_{n+1}(n+1)+a_{n}\right] x^{n}=0
\end{aligned}
$$

More importantly, we'll need:

$$
a_{n+1}(n+1)+a_{n}=0
$$

We can now set up a *recurrence relation* - a formula to get the next coefficient in the series from the previous one:

$$
a_{n+1}=-\frac{a_{n}}{(n+1)} \quad n=0,1,2, \ldots
$$

in other words, if we can figure out $a_0$ from the boundary conditions, we can go step by step to get all the coefficients.

Let's generate a few:

$$
a_{1}=-\frac{a_{0}}{0+1}=-a_{0}
$$

$$
a_{2}=-\frac{a_{1}}{1+1}=-\frac{1}{2} a_{1}=-\frac{1}{2}\left(-a_{0}\right)=\frac{1}{2} a_{0}
$$

$$
a_{3}=-\frac{a_{2}}{2+1}=-\frac{1}{3} a_{2}=-\frac{1}{3}\left(\frac{1}{2} a_{0}\right)=-\frac{1}{3 \cdot 2} a_{0}
$$

Hmm.

$$
a_{n}=(-1)^{n} \frac{1}{n !} a_{0} \quad n=0,1,2, \ldots
$$

This means our series looks like:

$$
y(x)=\sum_{n=0}^{\infty}(-1)^{n} \frac{1}{n !} a_{0} x^{n}=a_{0} \sum_{n=0}^{\infty}(-1)^{n} \frac{1}{n !} x^{n}
$$

Which is, of course the power series for:

$$
e^{-x}=\sum_{n=0}^{\infty}(-1)^{n} \frac{1}{n !} x^{n}
$$


<div class="alert alert-block alert-success">

We can solve for many kinds of ordinary differential equations by use of power series. We generate new series for first and second derivatives, multiply coefficients, etc. and then ..
    
***very, very carefully*** consolidate the series and work out recurrence relations for the coefficients
    
</div>


### Second Order Differential Equations with Power Series

Let's try a second-order differential equation with power series:

$$
\frac{d^{2}}{d x^{2}} y(x)+y(x)=0
$$

Again, you can solve this using other methods.. **HINT: What is the solution?**

We'll use a power series as our *ansatz*

$$
y(x)=\sum_{n=0}^{\infty} a_{n} x^{n}
$$

We can plug it into the equation - but we need the second derivative:

$$
\frac{d^{2}}{d x^{2}} y(x)=\frac{d^{2}}{d x^{2}} \sum_{n=0}^{\infty} a_{n} x^{n}=\sum_{n=0}^{\infty} a_{n}\left(\frac{d^{2}}{d x^{2}} x^{n}\right)=\sum_{n=0}^{\infty} a_{n} n(n-1) x^{n-2}
$$

So our differential equation is now:

$$
\frac{d^{2}}{d x^{2}} y(x)+y(x)=\sum_{n=0}^{\infty} a_{n} n(n-1) x^{n-2}+\sum_{n=0}^{\infty} a_{n} x^{n}=0
$$

Much like with the first-order equation, this isn't all that helpful since the second derivative term has a different exponent $x^{n-2}$ vs. the original function $x^n$

Let's write out the first few terms:

$$
a_0 \times 0 \times (-1) \times x^{-2} + a_1 \times 1 \times 0 \times x^{-1} + a_2 \times 2 \times 1 \times x^{0} + \ldots
$$

So we can drop the first two terms because they're zero, and rewrite the series:

$$
\sum_{n=2}^{\infty} a_{n} n(n-1) x^{n-2} \rightarrow \sum_{n=0}^{\infty} a_{n+2}(n+2)(n+1) x^{n}
$$

So we can now combine the two series to get our answer:

$$
\begin{aligned}
\frac{d^{2}}{d x^{2}} y(x)+y(x) &=\sum_{n=0}^{\infty} a_{n+2}(n+2)(n+1) x^{n}+\sum_{n=0}^{\infty} a_{n} x^{n} \\
&=\sum_{n=0}^{\infty}\left[a_{n+2}(n+2)(n+1)+a_{n}\right] x^{n}=0
\end{aligned}
$$

Again, we set up a recurrence relation for the coefficients:

$$
a_{n+2}(n+2)(n+1)+a_{n}=0
$$

so solving in terms of $a_n$

$$
a_{n+2}=-\frac{1}{(n+2)(n+1)} a_{n} \quad n=0,1,2, \ldots
$$

Notice that this time, we need two coefficients ($a_0$ and $a_1$) to crank through all of them:

$$
\begin{aligned}
&a_{0}\\
&a_{2}=-\frac{a_{0}}{2 \cdot 1} \quad a_{4}=-\frac{a_{2}}{4 \cdot 3}=\frac{a_{0}}{4 !} \quad a_{6}=-\frac{a_{4}}{6 \cdot 5}=-\frac{a_{0}}{6 !}
\end{aligned}
$$

So for the even coefficients:

$$
a_{2 n}=(-1)^{n} \frac{1}{(2 n) !} a_{0} \quad n=0,1,2, \ldots
$$

Then:

$$
a_{1} \quad a_{3}=-\frac{a_{1}}{3 \cdot 2} \quad a_{5}=-\frac{a_{3}}{5 \cdot 4}=\frac{a_{1}}{5 !} \quad a_{7}=-\frac{a_{5}}{7 \cdot 6}=-\frac{a_{1}}{7 !}
$$

All odd terms:

$$
a_{2 n+1}=(-1)^{n} \frac{1}{(2 n+1) !} a_{1} \quad n=0,1,2, \ldots
$$

Not that it's critical, but the final solution would look something like this:

$$
y(x)=a_{0} \sum_{n=0}^{\infty}(-1)^{n} \frac{1}{(2 n) !} x^{2 n}+a_{1} \sum_{n=0}^{\infty}(-1)^{n} \frac{1}{(2 n+1) !} x^{2 n+1}
$$

The two series look like:

$$
y(x)=a_{0} \cos (x)+a_{1} \sin (x)
$$

Again, that's not critical - we can use a few terms of the power series to approximate the solution at any point.
