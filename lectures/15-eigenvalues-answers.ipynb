### Q 15.4:
    
Consider the matrix:
    
$$
\mathbf{A}=\left(\begin{array}{rr}
-1 & 2 \\
2 & 2
\end{array}\right)
$$

(a) Determine the eigenvalues of A by finding the roots, $\lambda_1$ and $\lambda_2$, of the characteristic equation:

$$
|\mathbf{A}-\lambda \mathbf{I}|=0
$$

**Answer:**
    
Let's construct the matrix first:

$$
A - \lambda I
$$

$$
\left(\begin{array}{rr}
-1 & 2 \\
2 & 2
\end{array}\right) - \lambda
\left(\begin{array}{rr}
1 & 0 \\
0 & 1
\end{array}\right)
$$

$$
\left(\begin{array}{rr}
-1-\lambda & 2 \\
2 & 2-\lambda
\end{array}\right)
$$

So the determinant is:

$$
(-1 - \lambda)(2-\lambda) - 4 = (\lambda ^2 - \lambda -2) - 4 = \lambda ^2 - \lambda -6 = 0
$$

The roots are:

$$
\lambda = -2, +3
$$

```python
import numpy as np
from numpy.linalg import eig
```

```python
A = np.array([[-1, 2],
              [2, 2]])
e, v = eig(A)

print(e)
```

### Q 15.5

Consider the matrix:

$$
\mathbf{A}=\left(\begin{array}{ll}
1 & 1 \\
1 & 1
\end{array}\right)
$$

(a) Determine the eigenvalues of A by finding the roots, $\lambda_1$ and $\lambda_2$, of the characteristic equation:

$$
|\mathbf{A}-\lambda \mathbf{I}|=0
$$

Answer:

Much like before, we construct the matrix and find the determinant:

$$
\mathbf{A}=\left(\begin{array}{ll}
1-\lambda & 1 \\
1 & 1 - \lambda
\end{array}\right)
$$

The determinant is then:

$$
(1 - \lambda)(1-\lambda) - 1 = \lambda^2 - 2 \lambda + 1 - 1 = \lambda(\lambda - 2) = 0
$$

The roots are obviously 0 and 2.

```python
A = np.array([[1, 1],
              [1, 1]])
e, v = eig(A)
print(e)
```

### Q 15.6

Consider the matrix:

$$
\mathbf{A}=\left(\begin{array}{lll}
1 & 0 & 1 \\
0 & 1 & 0 \\
1 & 0 & 1
\end{array}\right)
$$

(a) Determine the eigenvalues of A by finding the roots, $\lambda_1$ and $\lambda_2$, of the characteristic equation:

$$
|\mathbf{A}-\lambda \mathbf{I}|=0
$$

Answer:
    
We construct first the matrix and then the determinant:

$$
\mathbf{A}=\left(\begin{array}{lll}
1-\lambda & 0 & 1 \\
0 & 1-\lambda & 0 \\
1 & 0 & 1-\lambda
\end{array}\right)
$$

Then the $3\times 3$ determinant is:

$$
(1-\lambda) \times ((1-\lambda)^2 - 0\times 0) - 0 \times (0\times (1-\lambda) - 0\times 1) + 1 \times(0\times 0 - 1\times (1-\lambda))
$$

$$
(1-\lambda)^3 - (1-\lambda) = 0
$$

$$
-\lambda^{3}+3 \lambda^{2}-2 \lambda = \lambda(-\lambda^2 + 3\lambda - 2) = 0
$$

One root is clearly 0, then solve the quadratic:

$$
-\lambda^2 + 3\lambda - 2 = 0
$$

So all three roots are 0, 1, 2

```python
A = np.array([[1, 0, 1],
              [0, 1, 0],
              [1, 0, 1]])
e, v = eig(A)
print(e)
```
